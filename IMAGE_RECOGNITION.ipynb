```python
import os
import random
from PIL import Image
import matplotlib.pyplot as plt
```

```python
images = ['n01630670', 'n01945685', 'n11901294']

names = {'n01630670': 'common newt', 'n01945685': 'slug', 'n11901294': 'iceland poppy'}

paths = []
```

```python
for folder in images:
    image_dir = os.path.join(r"C:\Users\Sindhu\Documents\New folder (2)\OneDrive\Desktop", folder)
    image_files = os.listdir(image_dir)
    paths.extend([(os.path.join(image_dir, file), names[folder]) for file in image_files])
```

```python
random_images = random.sample(paths, 10)
```

```python
for image_path, class_name in random_images:
    image_name = os.path.basename(image_path)
    image = Image.open(image_path)
    plt.imshow(image)
    plt.title(f"Image: {image_name}, Class: {class_name}")
    plt.axis('off') 
    plt.show()
```
